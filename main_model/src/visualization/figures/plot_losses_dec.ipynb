```python
import wandb
import pandas as pd

# Configurable list of run names
run_names = ["decoder_2d_redone_small", "decoder_torus_small", "decoder_sphere"]

# W&B project info
entity = "dagerber-massachusetts-institute-of-technology"
project = "decoder_training"

# Metrics to extract (one CSV per metric)
metrics_of_interest = [
    "train/loss",
    "train/feasibility_loss",
    "train/combined_loss"
]

# Initialize W&B API
api = wandb.Api()

# Loop over each metric separately
for metric in metrics_of_interest:
    metric_dataframes = []

    for run_name in run_names:
        # Find the run by display name
        runs = api.runs(f"{entity}/{project}", filters={"display_name": run_name})

        if not runs:
            print(f"Run '{run_name}' not found. Skipping.")
            continue

        run = runs[0]

        # Use scan_history() to ensure full history is retrieved
        rows = []
        for row in run.scan_history():
            if metric in row and "_step" in row:
                rows.append({
                    "step": row["_step"],
                    run_name: row[metric]
                })

        if not rows:
            print(f"No data found for '{metric}' in run '{run_name}'. Skipping.")
            continue

        df = pd.DataFrame(rows)
        metric_dataframes.append(df)

    # Merge all runs on step
    if metric_dataframes:
        merged_df = metric_dataframes[0]
        for df in metric_dataframes[1:]:
            merged_df = pd.merge(merged_df, df, on="step", how="outer")

        # Sort by step
        merged_df = merged_df.sort_values("step")

        # Save to CSV
        safe_metric_name = metric.replace("/", "_")
        csv_filename = f"{safe_metric_name}_merged.csv"
        merged_df.to_csv(csv_filename, index=False)
        print(f"Saved: {csv_filename}")
    else:
        print(f"No valid data found for metric: {metric}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator

def plot_data(
    csv_file, columns, steps,
    xlabel_epochs="Epochs", xlabel_steps="Steps", ylabel="Loss",
    log_scale=False, output='plot.png', percent_y=False, steps_per_epoch=391
):
    # Define RGB colors
    color_list = [
        (0.96, 0, 0.42),       # pink
        (0.216, 0.522, 0.882), # blue
        (0.22, 0.827, 0.835)   # cyan
    ]

    df = pd.read_csv(csv_file, usecols=columns.keys())
    df.rename(columns=columns, inplace=True)
    columns = list(columns.values())

    for col in columns[1:]:
        df[f'{col}_avg'] = df[col].rolling(steps, min_periods=1).mean()

    fig, ax = plt.subplots(figsize=(10, 6))
    x = df[columns[0]]

    for idx, col in enumerate(columns[1:]):
        color = color_list[idx % len(color_list)]
        ax.plot(x, df[col], color=color, alpha=0.45, linewidth=1, linestyle='dotted', label=f'{col} -- Raw', zorder=1)
        ax.plot(x, df[f'{col}_avg'], color=color, alpha=1.0, linewidth=2, label=f'{col} -- Average', zorder=2)

    ax.set_ylabel(ylabel)

    if log_scale:
        ax.set_yscale('log')
    if percent_y:
        ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: r"${:.0f}\%$".format(y)))

    # Space out the ticks with 25 * 391
    tick_interval = 25 * steps_per_epoch
    ax.set_xticks(x[::tick_interval].tolist() + [x.iloc[-1]])
    ax.set_xticklabels([])  # Hide the labels for the main x-axis
    ax.set_xlabel("")

    # Bottom x-axis: Epochs
    def step_to_epoch(step): return step / steps_per_epoch
    def epoch_to_step(epoch): return epoch * steps_per_epoch

    secax_bottom = ax.secondary_xaxis('bottom', functions=(step_to_epoch, epoch_to_step))
    secax_bottom.set_xlabel(xlabel_epochs)
    secax_bottom.xaxis.set_major_locator(MaxNLocator(integer=True))  # Integer ticks

    # Top x-axis: Steps
    secax_top = ax.secondary_xaxis('top')
    secax_top.set_xlabel(xlabel_steps)

    # Add grid lines for both axes
    ax.grid(True, which="both", ls="--", linewidth=0.5, alpha=0.7)  # Main axis grid

    # Add the second line of grid lines every 10,000 steps
    ax.grid(True, which='both', axis='x', linestyle=':', linewidth=0.5, alpha=0.7, color='gray')  # Secondary grid lines
    ax.set_xticks([10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000], minor=True)  # Add minor ticks
    ax.grid(True, which='minor', axis='x', linestyle=':', linewidth=0.5, alpha=0.7, color='gray')  # Minor grid lines

    ax.legend()

    plt.tight_layout()
    plt.savefig(output, format='pdf', transparent=True, bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
losses = ["combined_loss", "feasibility_loss", "loss"]
losses_latex = [
    r"$L_{\mathrm{decode}} \ (\log_{10})$",
    r"$L_{\mathrm{node}} \ (\log_{10})$",
    r"$L_{\mathrm{feas}} \ (\log_{10})$"
]
# steps = [100]
for loss, loss_latex in zip(losses, losses_latex):
    columns = {
        'step': 'Step',
        f'decoder_2d_redone_small': '2D Plane',
        f'decoder_torus_small': 'Torus',
        f'decoder_sphere': 'Sphere'
    }
    plot_data(
        csv_file=f'csv_files/train_{loss}_merged.csv',
        columns=columns,
        steps=500,
        xlabel_epochs='Epoch',
        xlabel_steps='Step',
        ylabel=f'{loss_latex}',
        log_scale=True,
        output=f'losses_decoder/{loss}.pdf',
        percent_y=False
    )
```
